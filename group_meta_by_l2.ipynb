```python
import pandas as pd
import os
```

```python
df_meta = pd.read_csv('/data2st1/junyi/output/atac0627/3REGIONS_peak_final.obs.csv',index_col=0)
```

```python
outdir = '/data1st2/junyi/output/atac0627/tobiasbam'
```

```python
df_meta['barcode'] = df_meta.index.str.split(':').str[1]  # 去掉条形码中的'-'部分
```

```python
for sample in df_meta['sample'].unique():
    if os.path.exists(os.path.join(outdir, sample)):
        print(f"Sample {sample} already exists in {outdir}. Skipping.")
    else:
        os.makedirs(os.path.join(outdir, sample), exist_ok=True)
    dfmeta_sub = df_meta.loc[df_meta['sample']==sample,['sample','barcode','celltype.L2.Condition']]
    for celltype in dfmeta_sub['celltype.L2.Condition'].unique():
        dfmeta_sub2 = dfmeta_sub[dfmeta_sub['celltype.L2.Condition'] == celltype]
        dfmeta_sub2['barcode'].to_csv(os.path.join(outdir, sample, f"{celltype}.csv"),header=False, index=False)
        print(f"Saved {celltype} metadata for sample {sample} to {outdir}/{sample}/{celltype}.csv")
```
