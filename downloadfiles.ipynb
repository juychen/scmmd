```python
PFC_AMY_subregions = ['MOs','PL.ORB','PL,ACA,ILA','CTXsp,sAMY']
selected_cells = meta_data[(meta_data.SubRegion.isin(PFC_AMY_subregions)) | (meta_data.MajorRegion.isin(["HPF"]))].index
selected_meta = meta_data.loc[selected_cells]
```

```python
sample_selected = selected_meta.merge(meta_sample, left_on='cell', right_on='Sample_title')
```

```python
values_download = sample_selected.Sample_supplementary_file_1.values
```

```python
length = len(values_download)
parts = 10
part_length = int(length/parts)
parts_list = [values_download[i*part_length:(i+1)*part_length] for i in range(parts)]
```

```python
if length % parts != 0:
    parts_list.append(values_download[parts*part_length:])
```

```python
sumlenglth = 0
for pl in parts_list:
    sumlenglth += len(pl)
if sumlenglth != length:
    print("error")
```

```python
for i, part in enumerate(parts_list):
    filename = f"/data2st1/junyi/methlyatlas/mCseq/allcfiles/part_{i+1}.txt"
    with open(filename, 'w') as file:
        for element in part:
            file.write(f"{element}\n")
    print(f"Part {i + 1} written to {filename}")
```
