```python
for keys in results_dict.keys():
    celltype = keys.split('/')[-1].replace('.tsv.gz','')
    array_results = np.array(results_dict[keys])
    df_top['mc_'+celltype] = array_results[:,0]
    df_top['cov_'+celltype] = array_results[:,1]
```

```python
df_top.columns
```

```python
#df_top['bulk_mc'] =  
df_top['counth'] = df_top['couontb'].str.split(r'[:,]').str[1].astype(int)
df_top['countm'] = df_top['couontb'].str.split(r'[:,]').str[3].astype(int)
```

```python
df_top['counthm'] = df_top['countm'] + df_top['counth']
```

```python
total_cov = df_top.filter(like='cov').sum(axis=1).values

for celltype in cell_types:
    df_top['counthm'] = df_top['counthm'].astype(int)
    df_top['frac_'+celltype] = df_top['mc_'+celltype].astype(float)/df_top['cov_'+celltype]
    df_top['pred_'+celltype] = df_top['mc_'+celltype].astype(float)/total_cov * df_top['counthm'].astype(float)
```

```python
total_frac = df_top.filter(like='frac_').sum(axis=1).values
for celltype in cell_types:
    df_top['enrich_'+celltype] = df_top['frac_'+celltype].astype(float)/total_frac.astype(float)
```
