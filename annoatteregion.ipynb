```python
def bb_to_bed(bb_file, bed_file):
    chrom_sizes = bb_file.chroms()
    
    with open(bed_file, "w") as outfile:
        for chrom, size in chrom_sizes.items():
            entries = bb_file.entries(chrom, 0, size)
            for entry in entries:
                start, end, name = entry[:3]  # Extract start, end, and name
                name_parts = name.split("\t")
                if "PLS" in name_parts[6]:
                    continue
                else:
                    value = name_parts[6]

                outfile.write(f"{chrom}\t{start}\t{end}\t.\t.\t{value}\n")
bb_to_bed(bb_file, f'output/convert_enhancer.bed')
```

```python
def merge_bed_files(bed_file1, bed_file2, output_file):
    # Load BED files
    bed1 = BedTool(bed_file1)
    bed2 = BedTool(bed_file2)

    # Intersect and split regions
    # Retain annotations from bed_file2 for overlapping regions
    merged = bed1.intersect(bed2, wa=True, wb=True).saveas("output/temp_intersect.bed")

    # Split non-overlapping regions from bed_file1
    non_overlapping = bed1.subtract(bed2).saveas("output/temp_non_overlapping.bed")
    
    # Split non-overlapping regions from bed_file1
    df_bed_1 = pd.read_csv('output/temp_intersect.bed',sep='\t',header=None)
    df_bed_2 = pd.read_csv('output/temp_non_overlapping.bed',sep='\t',header=None)

    def sort_key(chromosome):
        if chromosome.startswith('chr'):
            # Extract the numeric part for sorting
            return (0, int(chromosome[3:])) if chromosome[3:].isdigit() else (1, chromosome)
        return (2, chromosome)  # For any other cases
    
    df_bed_1=df_bed_1.iloc[:,6:]
    df_bed_1.columns = ['chromosome','start','end','name','score','annotation']
    df_bed_2.columns = ['chromosome','start','end','name','score','annotation']

    df_cat = pd.concat([df_bed_1,df_bed_2]).sort_values(by=["start","end"])
    df_cat['sort_key'] = df_cat['chromosome'].apply(sort_key)
    df_sorted = df_cat.sort_values(by=['sort_key', 'start']).drop(columns='sort_key')
    df_sorted=df_sorted[df_sorted.start!=df_sorted.end]
    df_sorted.to_csv(output_file,sep='\t',header=None,index=None)

merge_bed_files("/data2st1/junyi/generegion_vM33/intergenic.bed", "output/enhancer_m39converted.bed", "output/merged_output.bed")
```

```python
def check_bed_file(bed_file):
    """
    Check if a BED file contains any regions where start > end.
    """
    with open(bed_file, "r") as file:
        for line_number, line in enumerate(file, start=1):
            # Skip comment lines (if any)
            if line.startswith("#") or line.strip() == "":
                continue
            
            # Split the line into fields
            fields = line.strip().split("\t")
            
            # Ensure the line has at least 3 fields (chrom, start, end)
            if len(fields) < 3:
                print(f"Warning: Line {line_number} has fewer than 3 fields: {line.strip()}")
                continue
            
            # Extract chrom, start, and end
            chrom, start, end = fields[0], fields[1], fields[2]
            
            # Convert start and end to integers
            try:
                start = int(start)
                end = int(end)
            except ValueError:
                print(f"Warning: Line {line_number} has non-integer start/end values: {line.strip()}")
                continue
            
            # Check if start > end
            if start > end:
                print(f"Error: Line {line_number} has start > end: {line.strip()}")
            
            # Optional: Check if start == end (zero-length regions)
            if start == end:
                print(f"Warning: Line {line_number} has start == end (zero-length region): {line.strip()}")

# Example usage
check_bed_file("output/merged_output.bed")
```
