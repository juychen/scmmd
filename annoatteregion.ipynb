```python
## Calculate the mean methylation level for each gene in each enhancer region
# %%
stats =[]
CGN_values = []
matching_annotations = []

#for chrom, length in bb_file.chroms().items():
for index,row in use_gene_meta.iterrows():
    # Fetch all entries for the chromosome
    chrom = row['chrom']
    start = row['start']
    end = row['end']
    flaking = 2000

    entries = bb_file.entries(chrom, start-flaking, end+flaking)
    
    # Filter entries for genes in the DataFrame
    if  entries is None:
        continue
    for entry in entries:
        start, end, name = entry[:3]  # Extract start, end, and name
        values = bw.values(chrom, start, end)
        arr_values = np.array(values)
        arr_values = np.nan_to_num(arr_values)
        stats.append(arr_values.mean())
        CGN_values.append(arr_values)
        matching_annotations.append(name)
```

```python
maxlength = max([len(x) for x in CGN_values])
CGN_values = [np.pad(x,(0,maxlength-len(x))) for x in CGN_values]
```

```python
enhancer_methylation = np.array(CGN_values)
```

```python


if not os.path.exists(f'output/enhancer_meth/{cell_group}'):
    os.makedirs(f'output/enhancer_meth/{cell_group}')
plt.hist(stats, bins=100, color='skyblue', edgecolor='black')
# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.savefig(f'output/enhancer_meth/{cell_group}/{methtype}.png')
plt.close()

# value_array = np.array(CGN_values)
# np.save(f'output/enhancer_meth/{cell_group}/{methtype}.npy',value_array)
# stats = np.array(stats)
# np.save(f'output/enhancer_meth/{cell_group}/{methtype}_stats.npy',stats)
np.save(f'output/enhancer_meth/{cell_group}/{methtype}_enhancer_meth.npy',enhancer_methylation)
```
